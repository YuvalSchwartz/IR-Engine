```python
from google.cloud import storage
import pickle
import os

client = storage.Client()
bucket_name = 'wikidata_208373274'
bucket = client.bucket(bucket_name)
```

## **Importing indexes bins**

```python
title_index_blobs = bucket.list_blobs(prefix = 'title_index/title_index_bins/')
for blob in title_index_blobs:
    if '.bin' in blob.name:
        dir_path = os.path.dirname('/home/dataproc/' + 'title_index/title_index_bins/')
        os.makedirs(dir_path, exist_ok=True)
        blob.download_to_filename('/home/dataproc/' + blob.name)
```
